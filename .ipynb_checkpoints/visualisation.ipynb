# Visualisation
- Show variable relationships within the dataset.
- Determine the important variables to use within movie filtering.
- Help distinguish anomalies in the dataset.

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression
```

```python
%store -r movies
```

# Visualisation 1 (Pie Chart) : Release_Status
- Show that all movies in our dataset are 100% released.
- No movies should be recommended that are not released in cinemas.
- This proves preprocessing has worked correctly for this column.

```python
# Ensure all Movies are Released
def printDistribution():
    status_counts = movies["status"].value_counts()
    # Plotting the pie chart
    plt.figure(figsize=(8, 6))
    plt.pie(status_counts, labels=status_counts.index, autopct='%1.1f%%', startangle=140)
    plt.title('Distribution of Movies by Release Status')
    plt.axis('equal') 
    plt.show()
printDistribution()
```

- Results suggest that all data in our dataframe is of the released category in ["release_date"]

# Visualisation 2 (Histogram) : Numerical Columns
 - Investigation on the frequencies or occurences between each column.
 - Suggesting a use of IQR or a movement through EDA 
 - This shows anomalies in the dataset and therefore outliers can be removed later on.
 - Aim to find noise in the dataset.

```python
def printHistogram():
    plt.figure(figsize=(12, 6))
    movies[['budget', 'revenue', 'runtime', 'popularity', 'vote_average', 'vote_count']].hist(bins=20, figsize=(12, 8))
    plt.suptitle('Histograms of Numerical Variables', y=0.95)
    plt.show()
printHistogram()
```

- Results suggest that there is a large range of results and anomalies in the following columns.
    - Vote-Average
        - Movies with a vote_average of 0 suggest that the movie has not been voted for.
    - Runtime
        - Movies with a runtime of 0 will be required to be removed as this column is not possible.
    - Vote-Count
        - Lots of movies have not been voted for suggesting a lower vote-average and therefore potentially newer/older movies that have not been viewed.

- Considerations:
    - IQR to remove non-voted movies from the dataset to reduce noise and make better.
    - Outlier detection to help preprocess the data set.

# Visualisation 6 (Scatter Plot) : Revenue vs Budget
- Investigate the linear relationship between Revenue and Budget.
- This investigation stems off from visualisation 3 (pairplot) and 4 (correlational matrix).
- This allows a deeper understanding of the relationship between the variables.

```python
def printBudgetRevenusScatter():
    x=movies['revenue']
    y=movies['budget']
    plt.scatter(x=x, y=y)
    
    slope, intercept, rvalue, pvalue, stderr = stats.linregress(x=x,y=y)
    plt.plot(x, slope*x+intercept, color='red')
    
    print("y=%.3f + %.3f \nR: %.3f \nP: %.3f"%(slope, intercept, rvalue, pvalue)) # Equation Line
    plt.show()
printBudgetRevenusScatter()
```

- Results Suggest:
    - There is a positive linear relationship generated by a linear equation between the variables.
    - Therefore as Revenue increases so does Budget.
    - This suggests a directly proportional relationship.

- Next Steps
    - Investigate whether this affects an important factor in filtering movies i.e., popularity.

# Visualisation 8 (3D Scatter Plot) : Revenue, Budget and Popularity
- Investigation of the relationship between all three variables in one linear regression model.

```python
# Prepare data
X = movies[['budget', 'revenue']]
y = movies['popularity']

# Fit linear regression
coefficients = np.linalg.lstsq(X, y, rcond=None)[0]

# Create 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot
ax.scatter(movies['budget'], movies['revenue'], movies['popularity'], c='b', marker='o')

# Plot regression plane
xx, yy = np.meshgrid(np.linspace(movies['budget'].min(), movies['budget'].max(), 10), 
                     np.linspace(movies['revenue'].min(), movies['revenue'].max(), 10))
zz = coefficients[0]*xx + coefficients[1]*yy
ax.plot_surface(xx, yy, zz, alpha=0.5)

ax.set_xlabel('Budget')
ax.set_ylabel('Revenue')
ax.set_zlabel('Popularity')

plt.show()
```

- Results Suggest:
    - There is a strong positive relationship shown by the regression plane between the three variables.
    - Therefore suggesting that these three variables will be useful when filtering movies to suggest to users.

# Visualisation 9 (Pie Chart) : Categorical 
- To investigate the distribution of movies dependent on languages that the movies are spoken in.
- This may determine which films could be filtered by if a person can understand a certain language.

```python
def printLanguageDistribution():
    # Assuming `movies` is your dataframe with a column 'spoken_languages'
    all_languages = [lang for sublist in movies["spoken_languages"] for lang in sublist]
    language_counts = pd.Series(all_languages).value_counts()

    # Filtering languages that constitute less than 2% of the total
    total_movies = len(all_languages)
    language_counts = language_counts[language_counts / total_movies >= 0.02]

    # Plotting the pie chart
    plt.figure(figsize=(8, 6))
    plt.pie(language_counts, labels=language_counts.index, autopct='%1.1f%%', startangle=140)
    plt.title('Distribution of Movies by Spoken Languages')
    plt.axis('equal')
    plt.show()

# Call the function
printLanguageDistribution()
```

Results Suggest:
- Films in the dataset are predominantly English. 
- Therefore suggesting that the dataset used along with a movie recommendation system may benefit those that understand the English language based on majority measures.
- This may influence the model used to recommend movies based on language of the user.

# Visualisation 10 (Pie Chart): Release_Date
- Understand the distribution of the release date of movies within the dataset.
- May be a large spread of older and newer movies within the set.

```python
def printReleaseDateDistribution():
    # Assuming `movies` is your dataframe with a column 'spoken_languages'
    all_dates = [date for sublist in movies["release_date"] for date in sublist]
    language_counts = pd.Series(all_dates).value_counts()

    # Filtering languages that constitute less than 2% of the total
    total_movies = len(all_dates)
    language_counts = language_counts[language_counts / total_movies >= 0.02]

    # Plotting the pie chart
    plt.figure(figsize=(8, 6))
    plt.pie(language_counts, labels=language_counts.index, autopct='%1.1f%%', startangle=140)
    plt.title('Distribution of Movies by Release Date')
    plt.axis('equal')
    plt.show()

# Call the function
printReleaseDateDistribution()
```

Results suggest:
- There is a widespread of movie release dates within the early 20th century.
- Less movies are displayed from the late 19th century.
